# **Imports **

```python
import numpy as np
import tensorflow as tf
import keras
from keras.applications.inception_resnet_v2 import InceptionResNetV2,decode_predictions,preprocess_input
from keras.preprocessing.image import img_to_array,load_img
from IPython.core.display import display
```

#Input

```python
File_1='01 Umbrella.jpg'
File_2='02 Couple.jpg'
File_3='03 Ocean.jpg'
```

# **Step wise Preprocessing**

```python
pic=load_img(File_1,target_size=(299,299))
```

```python
display(pic)
```

```python
pic_array=img_to_array(pic)
```

```python
pic_array.shape
```

```python
expanded=np.expand_dims(pic_array,axis=0)
expanded.shape
```

```python
preprocess_img=preprocess_input(expanded)
```

#creating function for pre processsing image

```python
def imageprocessor (filename):
  img =load_img(filename,target_size=(299,299))
  display(filename)
  img=img_to_array(img)
  img=np.expand_dims(img,axis=0)
  img=preprocess_input(img)
  return img
```

# ** Model Training**

```python
inc_model=InceptionResNetV2(weights='imagenet')
```

```python
predictions=inc_model.predict(preprocess_img)
decode_predictions(predictions)
```

```python
image=imageprocessor(File_1)
img=load_img(File_1)
display(img)
prediction2=inc_model.predict(image )
decode_predictions(prediction2)
```
